```python
import streamlit as st
from streamlit_calendar import calendar

st.set_page_config(page_title="Training Calendar", layout="wide")

st.title("📅 Employee Training Calendar")

# Calendar events
events = [
    {
        "title": "PM Basics",
        "start": "2026-07-07",
        "end": "2026-07-07",
        "color": "#28a745"
    },
    {
        "title": "Safety Training",
        "start": "2026-07-10",
        "end": "2026-07-10",
        "color": "#007bff"
    },
    {
        "title": "Hydraulic Systems",
        "start": "2026-07-15",
        "end": "2026-07-15",
        "color": "#ffc107"
    },
    {
        "title": "RCA Workshop",
        "start": "2026-07-22",
        "end": "2026-07-22",
        "color": "#dc3545"
    }
]

calendar_options = {
    "initialView": "dayGridMonth",
    "headerToolbar": {
        "left": "prev,next today",
        "center": "title",
        "right": "dayGridMonth,timeGridWeek,timeGridDay"
    },
    "editable": True,
    "selectable": True,
    "navLinks": True,
    "height": 700
}

state = calendar(
    events=events,
    options=calendar_options,
    key="training_calendar",
)

st.write("### Selected Event")
st.write(state)
```
